## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List, Dict, Tuple
from collections import defaultdict, Counter
import re


class BPETokenizer:
    """
    Byte Pair Encoding (BPE) Tokenizer class.
    
    This tokenizer uses the BPE algorithm to iteratively merge the most frequent pairs of tokens/characters
    in a given corpus until the desired vocabulary size is reached.
    """

    def __init__(self, vocab_size: int) -> None:
        """
        Initializes the BPETokenizer with a specified vocabulary size.

        Args:
            vocab_size (int): The maximum size of the vocabulary.
        """
        self.vocab_size = vocab_size
        self.vocab = {}
        self.merge_rules = []

    def train(self, corpus: List[str]) -> None:
        """
        Trains the BPETokenizer on the given corpus.

        Args:
            corpus (List[str]): The corpus used for training.
        """
        # Count occurrences of all symbols and symbols pairs in the corpus
        vocab = Counter(" ".join(list(word) + ["</w"]) for word in corpus)

        # Determine the number of initial merges needed to reach the desired vocab size
        unique_characters = set(" ".join(corpus))
        initial_merges = self.vocab_size - len(unique_characters) - 1

        # Perform initial merges
        for _ in range(max(0, initial_merges)):
            pairs = self.get_stats(vocab)
            if not pairs:
                break
            best = max(pairs, key=pairs.get)
            if pairs[best] < 2:
                break
            vocab = self.merge_vocab(best, vocab)
            self.merge_rules.append(best)

        # Prune vocabulary to include only symbols that appear at least twice
        self.vocab = {k: v for k, v in vocab.items() if v >= 2}

    def get_stats(self, vocab: Dict[str, int]) -> Dict[Tuple[str, str], int]:
        """
        Get counts of pairs of consecutive symbols in the vocabulary.

        Args:
            vocab (Dict[str, int]): The vocabulary containing symbols and their frequencies.

        Returns:
            Dict[Tuple[str, str], int]: A dictionary containing counts of pairs of consecutive symbols.
        """
        pairs = defaultdict(int)
        for word, freq in vocab.items():
            symbols = word.split()
            for i in range(len(symbols)-1):
                pairs[symbols[i], symbols[i+1]] += freq
        return pairs

    def merge_vocab(self, pair: Tuple[str, str], v_in: Dict[str, int]) -> Dict[str, int]:
        """
        Merge a pair of symbols to produce a new vocabulary.

        Args:
            pair (Tuple[str, str]): The pair of symbols to merge.
            v_in (Dict[str, int]): The input vocabulary.

        Returns:
            Dict[str, int]: The output vocabulary after merging the pair.
        """
        v_out = {}
        bigram = re.escape(' '.join(pair))
        p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
        for word in v_in:
            w_out = p.sub(''.join(pair), word)
            v_out[w_out] = v_in[word]
        return v_out

    def tokenize(self, text: str) -> List[str]:
        """
        Tokenize a text using the trained BPE tokenizer.

        Args:
            text (str): The text to tokenize.

        Returns:
            List[str]: A list of tokens from the text.
        """
        word = ' '.join(text) + ' </w>'
        for pair in self.merge_rules:
            word = re.sub(re.escape(' '.join(pair)), ''.join(pair), word)
        return word.split()
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
text = []
merged_text = ""
files = []
for i in range(1,6):
    merged_text += open(f"./Assignment1_data/file{i}").read()
    files.append(open(f"./Assignment1_data/file{i}").read().split('\n')[:-1])
    for tweet in files[i-1]:
        for word in tweet.split():
            text.append(word)
```

### Train tokenizer with dfifferent vocab sizes

```python
vocab_sizes = [100, 250, 500, 750, 1000]
avg_lengths = []

for vocab_size in vocab_sizes:
    tokenizer = BPETokenizer(vocab_size=vocab_size)
    tokenizer.train(text)
    lengths = []
    for file in files:
        for tweet in file:
            # Add length of the tokenized tweet to the list of tokenized tweet lengths 
            lengths.append(len(tokenizer.tokenize(tweet)))
    
    # Add tweet average length for them to be plotted later
    avg_lengths.append(sum(lengths) / len(lengths))
```

### Plotting the results

```python
from matplotlib import pyplot as plt

plt.plot(vocab_sizes, avg_lengths)
plt.xlabel('Vocabulary Size')
plt.ylabel('Average Tweet Length (by tokens)')
plt.title("Average tweet length VS Vocabulary size")
plt.show()
```

### <span style="color:yellow">Answering the question above:</span>. 

In analyzing the generated graph above, which depicts the relationship between vocabulary size and the average length of tokenized tweets, we observe a clear inverse correlation: as the vocabulary size increases, the average tweet length, measured in tokens, decreases.

This trend is aligned with the core principle of Byte-Pair Encoding (BPE): a larger vocabulary size allows for more common sequences of characters (and subsequently sequences of characters that make up words) to be merged into single tokens. Consequently, a single token can represent longer sequences or entire words, which means that fewer tokens are needed to represent a tweet as the vocabulary size grows.

The preferable vocabulary size would strike a balance between granularity and compression. A smaller vocabulary size may not sufficiently capture common word fragments, leading to a less efficient representation, while an excessively large vocabulary size could overfit to the training corpus, incorporating infrequent or unique sequences that do not generalize well.

In practice, the choice of vocabulary size may be influenced by the following considerations:

- Computational Efficiency: Larger vocabularies may lead to slower processing times during tokenization and training of downstream models.
- Generalization: A moderately sized vocabulary may generalize better to unseen text than an overly large one, which might be too tailored to the training data.
- Downstream Task Performance: The impact on the performance of any machine learning models using these tokens for tasks like classification, sentiment analysis, etc., should be considered. The vocabulary size that leads to the best performance on validation data is often preferred.

From the graph, we can deduce that a point of diminishing returns may be reached; beyond a certain vocabulary size, the benefit of additional tokens may not justify the costs. The ideal vocabulary size is therefore dataset and task-specific, and it should be chosen based on empirical results from experiments on validation sets relevant to the intended application.

## Solution:

I'm using a dataset in Italian that I found on github [here](https://github.com/charlesmalafosse/open-dataset-for-sentiment-analysis/blob/master/betsentiment-IT-tweets-sentiment-teams-split.zip.002). As the dataset very large in size, I'm using only the first 0.1% of the rows that the .csv file contains.

```python
import pandas as pd

# Read the dataset
data = pd.read_csv(
    "./Assignment1_data/betsentiment-IT-tweets-sentiment-players.csv", encoding='latin1')

# Get the first 0.1% of the rows and extract the "tweet_text" column and convert it to a list
tweet_texts = data.head(int(len(data) * 0.01))["tweet_text"].tolist()
```

```python
it_vocab_sizes = [100, 250, 500, 750, 1000]
it_avg_lengths = []

for it_vocab_size in it_vocab_sizes:
    it_tokenizer = BPETokenizer(vocab_size=it_vocab_size)
    it_tokenizer.train(tweet_texts)
    lengths = []
    for it_tweet in tweet_texts:
        tokenized_tweet = it_tokenizer.tokenize(it_tweet)
        lengths.append(len(tokenized_tweet))

    avg_tweet_len = sum(lengths) / len(lengths)
    it_avg_lengths.append(avg_tweet_len)
```

```python
from matplotlib import pyplot as plt

plt.plot(it_vocab_sizes, it_avg_lengths)
plt.xlabel('Vocabulary Size')
plt.ylabel('Average Tweet Length (by tokens)')
plt.title("Average tweet length VS Vocabulary size IN ITALIAN")
plt.show()
```
